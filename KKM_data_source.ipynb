## KKM Dataset

```python
import pandas as pd
import numpy as np
import requests as rs
import json
import warnings
warnings.filterwarnings("ignore")
```

```python
# Import data from KKM source in Google Sheets

csv_url='https://docs.google.com/spreadsheets/d/15A43eb68LT7gg_k9VavYwS1R2KkCTpvigYMn5KT9RTM/export?format=xlsx&id=15A43eb68LT7gg_k9VavYwS1R2KkCTpvigYMn5KT9RTM'
res=rs.get(url=csv_url)
xlsfile = 'rawfiles/Covid19-KKM.xlsx'
open(xlsfile, 'wb').write(res.content)
df = pd.read_excel(xlsfile, None)

wp = json.load(open('rawfiles/wilayah.json'))

# Available sheets
df.keys()
```

## Daily cases tables

```python
# Testing and results data 1

dfdbase = df['d(Base)'].set_index('Date')
dfdbase['Tested'] = dfdbase['Tested'].fillna(method='ffill')
display(dfdbase.columns)
display(dfdbase.tail())
#display(dfdbase.dtypes)
dfdbase.to_csv('rawfiles/{file}'.format(file='dbase.csv'))
```

```python
# Testing and results data 2

dfbase = df['Base'].set_index('Date').fillna(method='ffill')
for col in dfbase.columns:
    if 'Unnamed' in col: dfbase = dfbase.drop(col, axis=1)
display(dfbase.columns)
display(dfbase.tail())
#display(dfbase.dtypes)
dfbase.to_csv('rawfiles/{file}'.format(file='base.csv'))
```

```python
# Total confirmed cases by state

dfstate = df['State']
dfstate['WP Kuala Lumpur'] = dfstate.apply(lambda row: row['WP KL & Putrajaya'] if np.isnan(row['WP Kuala Lumpur']) else row['WP KL & Putrajaya'], axis=1)
dfstate = dfstate.fillna(method='ffill').iloc[:,:-3]
dfstate = pd.melt(dfstate, id_vars=['Date']).set_index('Date')
dfstate = dfstate.rename(columns={'variable':'state', 'value':'cases'})
display(dfstate.tail(10))
dfstate.to_csv('rawfiles/{file}'.format(file='state.csv'))
```

```python
# Total confirmed cases by district

dfdistrict = pd.read_excel(xlsfile, sheet_name='District', header=[0,1])
dfdistrict = pd.melt(dfdistrict, id_vars=[dfdistrict.columns[0]])
dfdistrict = dfdistrict.rename(columns={dfdistrict.columns[0]:'date', 'variable_0':'state',
                                        'variable_1':'district', 'value':'cases'})
dfdistrict = dfdistrict[dfdistrict.state!='TOTAL'].drop(0).set_index('date')
dfdistrict.state = dfdistrict.state.apply(lambda row: wp['wp'][row] if row in list(wp['wp'].keys()) else row)
display(dfdistrict[10:20])
dfdistrict.to_csv('rawfiles/{file}'.format(file='district.csv'))
```

```python
# Patient details - NOTE: data last updated 6 March

dfpatients = df['Patients'].iloc[:,:-1].set_index('Case')
for col in ['Date Confirmed','Date Discharged','Date of Death']:
    dfpatients[col] = dfpatients[col].apply(lambda row: pd.to_datetime(row, format='%d/%m/%Y') if type(row)==str else (row if pd.isnull(row) else (pd.to_datetime(str('{d}/{m}/{y}'.format(m=str(row)[5:7], d=str(row)[8:10], y=str(row)[:4])), format='%d/%m/%Y') if int(str(row)[8:10]) > 12 else pd.to_datetime(str('{d}/{m}/{y}'.format(d=str(row)[5:7], m=str(row)[8:10], y=str(row)[:4])), format='%d/%m/%Y'))))
dfpatients = dfpatients.dropna(subset=['Sex']).reset_index()
dfpatients.State = dfpatients.State.apply(lambda row: wp['wp'][row] if row in list(wp['wp'].keys()) else row)
display(dfpatients.columns)
display(dfpatients.tail())
display(dfpatients.dtypes)
dfpatients.to_csv('rawfiles/{file}'.format(file='patients.csv'))
```

```python
# Death cases by patient

dfdeaths = df['Death_cases'].set_index('ID')
dfdeaths['Date warded'] = dfdeaths['Date warded'].apply(lambda row: np.nan if row=='.' else row)
for col in ['Date passed away','Date warded']:
    dfdeaths[col] = dfdeaths[col].apply(lambda row: pd.to_datetime(row, format='%d/%m/%Y') if type(row)==str else (row if pd.isnull(row) else (pd.to_datetime(str('{d}/{m}/{y}'.format(m=str(row)[5:7], d=str(row)[8:10], y=str(row)[:4])), format='%d/%m/%Y') if int(str(row)[8:10]) > 12 else pd.to_datetime(str('{d}/{m}/{y}'.format(d=str(row)[5:7], m=str(row)[8:10], y=str(row)[:4])), format='%d/%m/%Y'))))
dfdeaths.State = dfdeaths.State.apply(lambda row: wp['wp'][row] if row in list(wp['wp'].keys()) else row)
display(dfdeaths.columns)
display(dfdeaths.tail())
display(dfdeaths.dtypes)
dfdeaths.to_csv('rawfiles/{file}'.format(file='deaths.csv'))
```

```python
# Tabligh cluster

dftabligh = df['Tabligh'].iloc[:15]
dftabligh.Negeri = dftabligh.Negeri.apply(lambda row: wp['wp'][row] if row in list(wp['wp'].keys()) else row)
dftabligh = dftabligh.drop(dftabligh.columns[0], axis=1).set_index('Negeri')
for col in dftabligh.columns: dftabligh[col] = dftabligh[col].astype(int)
display(dftabligh.columns)
display(dftabligh.head())
dftabligh.to_csv('rawfiles/{file}'.format(file='tabligh.csv'))
```

## Reference tables

```python
# State demographics by age group

demographics = df['Demographic Profile (AgeState)'][:16]
demographics['Negeri'] = demographics['Negeri'].apply(lambda row: row.title())
demographics.Negeri = demographics.Negeri.apply(lambda row: wp['wp'][row] if row in list(wp['wp'].keys()) else row)
demographics = demographics.set_index('Negeri')
display(demographics.columns)
display(demographics.head())
demographics.to_csv('rawfiles/{file}'.format(file='demographics.csv'))
```

```python
# Hospital statistics

hospitals = df['Covid19 Hospitals'][:-10]
hospitals.columns = hospitals.iloc[0]
hospitals = hospitals[hospitals['Nama hospital']!='Jumlah'].drop(0)
hospitals.Negeri = hospitals.Negeri.apply(lambda row: wp['wp'][row] if row in list(wp['wp'].keys()) else row)
for col in hospitals.columns[6:9]:
    hospitals[col] = hospitals[col].astype(int)   
for col in hospitals.columns[9:]:
    hospitals[col] = hospitals[col].astype(float)
display(hospitals.columns)
display(hospitals.tail())
display(hospitals.dtypes)
hospitals.to_csv('rawfiles/{file}'.format(file='hospitals.csv'))
```

```python
# Quarantine stations

qstations = df['Q Stations']
qstations.State = qstations.State.apply(lambda row: wp['wp'][row] if row in list(wp['wp'].keys()) else row)
display(qstations.head())
display(qstations.dtypes)
qstations.to_csv('rawfiles/{file}'.format(file='qstations.csv'))
```

```python
# New reports

news = df['News']
display(news.head())
display(news.dtypes)
news.to_csv('rawfiles/{file}'.format(file='news.csv'))
```
